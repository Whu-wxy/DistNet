```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator, FormatStrFormatter


font1 = {'family' : 'Times New Roman',  
'weight' : 'normal',  
'size'   : 9,  
} 

font2 = {'family' : 'Times New Roman',  
'weight' : 'normal',  
'size'   : 14,  
}  

figsize = 8, 8
plt.subplots(figsize=figsize, dpi=120)                                # 设定整张图片大小

x = [0.3, 0.4, 0.5, 0.6, 0.7, 0.8]

CTW_F = [78.8, 80.4, 80.4, 79.9, 76.4, 72.8]
CTW_P = [85.4, 85.8, 84.8, 82.7, 76.9, 71.8]
CTW_R = [73.2, 75.6, 76.5, 77.2, 75.9, 73.8]
CTW_NUM = [2629, 2704, 2768, 2864, 3028, 3156]

Total_F = [76.5, 78, 78.4, 79.1, 77.2, 75.3]
Total_P = [83.3, 83.7, 82.7, 82, 78.4, 74.7]
Total_R = [70.8, 73.1, 74.6, 76.4, 76.1, 75.8]
Total_NUM = [2092, 2160, 2236, 2314, 2410, 2524]

####### 表一
ax1 = plt.subplot(2, 2, 1)
plt.xlim(0.2, 0.9)
plt.ylim(70, 90) 
plt.title('(a)', fontsize=12)
plt.xlabel('${D}_{max}$')
plt.ylabel('F-score')
#第二条线
plt.plot(x, CTW_F, color='red', label='$CTW$', linewidth=0.8, marker='.')
plt.legend(loc='upper right', prop=font1, frameon=False)                                # 绘制图例，指定图例位置
#三
plt.plot(x, Total_F, color='blue', label='$Total$', linewidth=0.8, linestyle='--', marker='+')
plt.legend(loc='upper right', prop=font1, frameon=False)                                # 绘制图例，指定图例位置

for a, b in zip(x, CTW_F):
    c = b
    if a>0.6:
        c -= 1.5
    plt.text(a, c+0.2, b, ha='center', va='bottom')  #, fontsize=10
for a, b in zip(x, Total_F):
    c = b
    if a==0.6:
        c -= 1.5
    plt.text(a, c+0.2, b, ha='center', va='bottom')  #, fontsize=10
    
    
####### 表二
ax2 = plt.subplot(2, 2, 2)
plt.xlim(0.2, 0.9)
plt.ylim(70, 90)
plt.title('(b)', fontsize=12)
plt.xlabel('${D}_{max}$')
plt.ylabel('Precision')
#第二条线
plt.plot(x, CTW_P, color='red', label='$CTW$', linewidth=0.8, marker='.')
plt.legend(loc='upper right', prop=font1, frameon=False)                                # 绘制图例，指定图例位置
#三
plt.plot(x, Total_P, color='blue', label='$Total$', linewidth=0.8, linestyle='--', marker='+')
plt.legend(loc='upper right', prop=font1, frameon=False)                                # 绘制图例，指定图例位置

for a, b in zip(x, CTW_P):
    c = b
    if a>0.6:
        c -= 1.5
    plt.text(a, c+0.2, b, ha='center', va='bottom')  #, fontsize=10
for a, b in zip(x, Total_P):
    c = b
    if a==0.6:
        c -= 1.5
    plt.text(a, c+0.2, b, ha='center', va='bottom')  #, fontsize=10

####### 表三
ax3 = plt.subplot(2, 2, 3)
plt.xlim(0.2, 0.9)
plt.ylim(70, 90)
plt.title('(c)', fontsize=12)
plt.xlabel('${D}_{max}$')
plt.ylabel('Recall')
#第二条线
plt.plot(x, CTW_R, color='red', label='$CTW$', linewidth=0.8, marker='.')
plt.legend(loc='upper right', prop=font1, frameon=False)                                # 绘制图例，指定图例位置

#三
plt.plot(x, Total_R, color='blue', label='$Total$', linewidth=0.8, linestyle='--', marker='+')
plt.legend(loc='upper right', prop=font1, frameon=False)                                # 绘制图例，指定图例位置

for a, b in zip(x, CTW_R):
    c = b
    if a>0.6:
        c -= 1.5
    plt.text(a, c+0.2, b, ha='center', va='bottom')  #, fontsize=10
for a, b in zip(x, Total_R):
    c = b
    if a==0.6:
        c -= 1.5
    plt.text(a, c+0.2, b, ha='center', va='bottom')  #, fontsize=10
    
# ####### 表四
ax4 = plt.subplot(2, 2, 4)
plt.xlim(0.2, 1)
plt.ylim(2000, 3400)
plt.title('(d)', fontsize=12)
plt.xlabel('${D}_{max}$')
plt.ylabel('number of the detections')

#第二条线
plt.plot(x, CTW_NUM, color='red', label='$CTW$', linewidth=0.8, marker='.')
plt.legend(loc='upper right', prop=font1, frameon=False)                                # 绘制图例，指定图例位置
#三
plt.plot(x, Total_NUM, color='blue', label='$Total$', linewidth=0.8, linestyle='--', marker='+')
plt.legend(loc='upper right', prop=font1, frameon=False)                                # 绘制图例，指定图例位置

for a, b in zip(x, CTW_NUM):
    c = b
    if a==0.8:
        c -= 120
    plt.text(a, c+20, b, ha='center', va='bottom')  #, fontsize=10
for a, b in zip(x, Total_NUM):
    plt.text(a, b+20, b, ha='center', va='bottom')  #, fontsize=10
    
    

plt.tight_layout()

plt.savefig("../../line_chart.png", dpi=600)
plt.show()
```
